```python
for i in lnk:
    try:
    #     # time.sleep(2)
        driver.get(i)
        driver.implicitly_wait(3)
        driver.find_element("xpath","//button[@aria-label = 'Click to see more description']").click()

        title = driver.find_element("xpath","//h1[@class='t-24 t-bold jobs-unified-top-card__job-title']").text.replace('\n','').strip()

        try:
            company =driver.find_element("xpath","//span[@class='jobs-unified-top-card__company-name']/a").text.replace('\n','').strip()
        except:
            company =driver.find_element("xpath","//span[@class='jobs-unified-top-card__company-name']").text.replace('\n','').strip()

        

        location = driver.find_element("xpath","//SPAN[@CLASS='jobs-unified-top-card__bullet']").text.replace('\n','').strip()
        try:
            typ = driver.find_element("xpath","//SPAN[@CLASS='jobs-unified-top-card__workplace-type']").text.replace('\n','').strip()
        except:
            typ = driver.find_element("xpath","//li[@class='jobs-unified-top-card__job-insight']/span").text.replace('\n','').strip()



        try:
            employees = driver.find_element("xpath","//div[@class='t-14 mt5']/span").text.replace('\n','').strip() 
        except:
            employees = 'Null'


        try:
            followers = driver.find_element("xpath","//div[@class='t-14 mt5']/span[2]").text.replace('\n','').strip().replace('\n','').strip()
        except:
            # try:
            #     followers = soup2.find('div',class_='artdeco-entity-lockup__content ember-view flex-grow-1').text.replace('\n','').strip()
            # except:
            followers = 'Null'  

        try:
            applicant = driver.find_element("xpath","//span[@class='jobs-unified-top-card__subtitle-secondary-grouping t-black--light']//span[@class='jobs-unified-top-card__bullet']").text.replace('\n','').strip() 
        except:
            try:
                applicant = driver.find_element('span',"//span[@class='jobs-unified-top-card__subtitle-secondary-grouping t-black--light']").text.replace('\n','').strip()
            except:
                applicant = 'Null'

        try:
            exact_applicant = driver.find_element("xpath","//li[@class='jobs-unified-top-card__job-insight jobs-unified-top-card__job-insight--highlight']").text.replace('\n','').strip()
        except:
            exact_applicant = "Null"

        industry = driver.find_element("xpath","//div[@class='t-14 mt5']").text.replace('\n','').strip()

        involvement = driver.find_element("xpath","//li[@class='jobs-unified-top-card__job-insight']/span").text.replace('\n','').strip()
        jbd = []
        try:
            driver.implicitly_wait(5)
            jd = driver.find_elements("xpath","//div[@id='job-details']//ul//li")
            # print(len(jd))
            # print(jd)
            for j in range(len(jd)):
                jbd.append(jd[j].text.replace('\n','').strip())
        except:
            pass

        if len(jbd) < 1:
            txt = driver.find_element("xpath","//div[@id='job-details']/span").text.replace('\n','').strip()
            jbd.append(txt)
        
        driver.find_element("xpath","//button[@class='jobs-unified-top-card__job-insight-text-button']").click()
        skills = driver.find_element("xpath","//ul[@class='job-details-skill-match-status-list']").text.replace('\n','').strip()
        print(skills)
        




        rows = [title,company,location,typ,employees,followers,applicant,industry,involvement,jbd,skills]
        print(rows)
        
        


        with open(f"C:/Users/HP/OneDrive/Desktop/Completed Projects/LinkedIn_daily_data/LinkedIn_NEW.csv",'a+',encoding='UTF8' ,newline='') as f:
            writer = csv.writer(f)     
            writer.writerow(rows)



    except:
        print("This Link is not working","  ",i)
        pass
```
